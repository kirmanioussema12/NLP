```python
# Load the tagged words from the Brown corpus (news category) with Universal POS tags
brown_news_tagged = brown.tagged_words(categories='news', tagset='universal')
```

```python
# Extract all verbs from the tagged Brown news corpus
verbes = [word for word, tag in brown_news_tagged if tag == 'VERB']
```

```python
# Extract all nouns from the Brown corpus and find the most frequent ones
noms = [word for word, tag in brown_news_tagged if tag == 'NOUN']
freq_noms = Counter(noms)
```

```python
noms_les_plus_frequents = freq_noms.most_common(10)
print(noms_les_plus_frequents)
```

```python
# Create a dictionary mapping each word to its set of POS tags
dictionnaire_mots_etiquettes = defaultdict(set)
for word, tag in brown_news_tagged:
    dictionnaire_mots_etiquettes[word].add(tag)


for mot, etiquettes in list(dictionnaire_mots_etiquettes.items())[:10]:
    print(f"{mot}: {etiquettes}")
```

```python
# Import various taggers (Unigram, Bigram, Trigram) and Brill for tagging experiments
from nltk.tag import UnigramTagger, BigramTagger, TrigramTagger
from nltk.tag import brill, brill_trainer
from nltk.tag.hmm import HiddenMarkovModelTagger
from sklearn.model_selection import train_test_split
```

```python
brown_tagged_sents = brown.tagged_sents(tagset='universal')
```

```python
# Split the Brown corpus into training (90%) and testing (10%) sets
train_sents, test_sents = train_test_split(brown_tagged_sents, test_size=0.1)
```

```python
# Import various taggers (Unigram, Bigram, Trigram) and Brill for tagging experiments
unigram_tagger = UnigramTagger(train_sents)
unigram_accuracy = unigram_tagger.evaluate(test_sents)
print(f'Unigram Tagger Accuracy: {unigram_accuracy:.4f}')
```

```python
# Train a Bigram tagger (with Unigram backoff) and evaluate its accuracy
bigram_tagger = BigramTagger(train_sents, backoff=unigram_tagger)
bigram_accuracy = bigram_tagger.evaluate(test_sents)
print(f'Bigram Tagger Accuracy: {bigram_accuracy:.4f}')
```

```python
# Train a Trigram tagger (with Bigram backoff) and evaluate its accuracy
trigram_tagger = TrigramTagger(train_sents, backoff=bigram_tagger)
trigram_accuracy = trigram_tagger.evaluate(test_sents)
print(f'Trigram Tagger Accuracy: {trigram_accuracy:.4f}')
```

```python
# Train a Hidden Markov Model (HMM) tagger and evaluate its accuracy
hmm_tagger = HiddenMarkovModelTagger.train(train_sents)
hmm_accuracy = hmm_tagger.evaluate(test_sents)
print(f'HMM Tagger Accuracy (Viterbi): {hmm_accuracy:.4f}')
```

```python
# Train a Brill tagger using the Unigram tagger as the initial tagger
templates = brill.fntbl37()
trainer = brill_trainer.BrillTaggerTrainer(initial_tagger=unigram_tagger, templates=templates)
brill_tagger = trainer.train(train_sents)
brill_accuracy = brill_tagger.evaluate(test_sents)
print(f'Brill Tagger Accuracy: {brill_accuracy:.4f}')
```
